## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
raw_data = pd.read_csv('Categorical.csv')
raw_data
```

Remove the duplicate index column from the dataset. 

```python
data = raw_data.copy()
data = data.iloc[:,1:4]
data
```

```python
data['continent'].unique()
```

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data_mapped = data
data_mapped['continent'] = data_mapped['continent'].map({'North America':0, 'Asia':1, 'Africa':2, 'Europe':3, 'South America':4,
       'Oceania':5, 'Antarctica':6, 'Seven seas (open ocean)':7})
data_mapped
```

## Select the features

```python
x = data_mapped.iloc[:,:3]
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
```

## Clustering results

```python
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data_mapped.copy()
data_with_clusters['cluster'] = identified_clusters
data_with_clusters
```

## Plot the data

```python
plt.scatter(raw_data['Longitude'],raw_data['Latitude'],c=data_with_clusters['cluster'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
